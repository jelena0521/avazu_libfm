```python
import os
import pandas as pd

from gensim.corpora import WikiCorpus
from gensim.models import Word2Vec
from gensim.models.word2vec import LineSentence
import pandas as pd
import multiprocessing
import numpy as np
import pickle

```

```python
#以下对app_id,site_id,site_domain进行w2v,完全可以写个函数进行封装，这里就算了


#以下对app_id进行w2v
import pickle
temp = pickle.load(open('app_id_history.pkl', 'rb'))
print(type(temp))
```

```python
L=6
sentence = []
for user,user_app_id_history in temp.items():
    sentence.append(user_app_id_history)
#print(sentence[:6])
```

```python
model = Word2Vec(sentence, size=L, window=2, min_count=1, workers=multiprocessing.cpu_count())
```

```python
model['a37bf1e4']
```

```python
app_id=pickle.load(open('app_id.pkl', 'rb'))
```

```python
print(app_id)
```

```python
#对app_id进行w2v

values = set(app_id)
w2v = []
for i in values:
    a = [i]
    try:
       a.extend(model[i])
       w2v.append(a)
    except:
       print("word {} not in vocabulary".format(i))
out_df = pd.DataFrame(w2v)

#print(model.similar_by_word('50e219e0'))

name = ['app_id']
for i in range(L):
    name.append(name[0] + 'W' + str(i))
out_df.columns = name
out_df.to_csv( 'app_id_em.csv', index=False)

print(out_df.head(20))
```
